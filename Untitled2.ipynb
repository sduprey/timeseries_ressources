```python
N = 100
X = np.random.random((N, 2)) * 6 - 3 # uniformly distributed between (-3, +3)
Z = np.cos(2*X[:,0]) + np.cos(3*X[:,1])
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X, Z);
```

```python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=128, alpha=0., learning_rate_init=0.01)
model.fit(X, Z);
```

```python
### compute the coordinates of a grid for the square [-3,3]
####https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html
```

```python
line = np.linspace(-3, 3, 50)
xx, yy = np.meshgrid(line, line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
Xgrid.shape
```

```python
Zhat = model.predict(Xgrid)
Zforesthat = regr.predict(Xgrid)
```

```python
Zhat.shape
```

```python
Ztrue = np.cos(2*Xgrid[:,0]) +  np.cos(3*Xgrid[:,1])
```

```python
Ztrue.shape
```

```python
# Plot the prediction surface
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Zhat, linewidth=0.2, antialiased=True);
```

```python
# Plot the prediction surface
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Zforesthat, linewidth=0.2, antialiased=True);
```

```python
# Plot the prediction surface
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Ytrue, linewidth=0.2, antialiased=True);
```

```python
### compute the loss RMSE
def my_rmse(x,xhat):
    return np.sqrt(np.power((x-xhat),2).mean())

my_rmse(Ztrue, Zhat)
```

```python
my_rmse(Ztrue, Zforesthat)
```

```python
for my_extrapolation in [5,10,15]:
    print(f'extrapoling {my_extrapolation}')
    line = np.linspace(-my_extrapolation, my_extrapolation, 200)
    xx, yy = np.meshgrid(line, line)
    Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
    Xgrid.shape
    Zhat = model.predict(Xgrid)
    Zforesthat = regr.predict(Xgrid)
    # Plot the prediction surface
    fig = plt.figure(figsize=(20, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Zforesthat, linewidth=0.2, antialiased=True);
    fig = plt.figure(figsize=(20, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Zhat, linewidth=0.2, antialiased=True);

    Ztrue = np.cos(2*Xgrid[:,0]) +  np.cos(3*Xgrid[:,1])
    print(my_rmse(Ztrue, Zhat))
    print(my_rmse(Ztrue, Zforesthat))
```

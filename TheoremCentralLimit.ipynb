```python
T = 10000
for n in [10, 100, 1000, 10000]:
    tirages=np.random.randn(n,T)
    data_df =pd.DataFrame(data = tirages)
    mean_distribution_df = data_df.mean(axis=0)
    mean_distribution_df.hist(bins=100)
    
```

```python
T = 10000
for n in [10, 100, 1000, 10000]:
    tirages=np.random.randn(n,T)
    data_df =pd.DataFrame(data = tirages)
    mean_distribution_df = np.sqrt(n)*data_df.mean(axis=0)
    mean_distribution_df.hist(bins=100)
```

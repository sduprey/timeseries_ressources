```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of time steps
T = 1000

# initial price
P0 = 10

# drift
mu = 0.001

sigma =  0.01
```

```python
# last log price
last_p = np.log(P0)

log_returns = np.zeros(T)
prices = np.zeros(T)
for t in range(T):
  # sample a log return
  r = sigma * np.random.randn()

  # compute the new log price
  p = last_p + mu + r

  # store the return and price
  log_returns[t] = r + mu
  prices[t] = np.exp(p)

  # assign last_p
  last_p = p
```

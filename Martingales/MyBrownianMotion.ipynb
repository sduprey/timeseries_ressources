```python
import numpy as np
import pandas as pd
```

```python
Nobs = 100
Ntraj = 1000
iid_samples = np.random.randn(Nobs,Ntraj) + 0.5
data_df = pd.DataFrame(data=iid_samples)
data_df = data_df.cumsum()
data_df.plot()
```

```python
for Ntimes in [10,50,100,500,1000]: 
    Ntraj = 1000
    iid_samples = np.random.randn(Ntimes,Ntraj)
    data_df = pd.DataFrame(data=iid_samples)
    series_mean = np.sqrt(Ntimes)*data_df.mean()
    #series_mean = data_df.mean()
    series_mean.hist(bins=50)
```

```python
for Ntimes in [10,50,100,500,1000]: 
    Ntraj = 1000
    iid_samples = np.random.randn(Ntimes,Ntraj)+0.5
    data_df = pd.DataFrame(data=iid_samples)
    data_df = data_df.cumsum()
    series_mean = data_df.iloc[-1]
    series_mean.hist(bins=50)
```
